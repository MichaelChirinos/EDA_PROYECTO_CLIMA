```python
import pandas as pd
import numpy as np
```

```python
dfs = []
for year in range(1996,2023):
  df = pd.read_excel(f"/content/drive/MyDrive/Data Cleaning/data/imeca{year}.xls")
  dfs.append(df)
```

```python
df_imeca = pd.concat(dfs, ignore_index=True)
df_imeca
```

```python
df_imeca.info()
```

```python
df_imeca[-pd.isna(df_imeca).FECHA].FECHA
```

```python
#Combinando Fecha y FECHA
df_imeca.Fecha = df_imeca.Fecha.combine_first(df_imeca.FECHA)
```

```python
df_imeca.info()
```

```python
#Aplicando a columnas restantes
zonas = ["Noroeste", "Noreste", "Centro", "Suroeste","Sureste"]
for zona in zonas:
  df_imeca[f"{zona} Ozono"] = df_imeca[f"{zona} Ozono"].combine_first(df_imeca[f"{zona} ozono"])
```

```python
df_imeca.info()
```

```python
#Borrando las columnas sobrantes
df_imeca = df_imeca.drop(columns=["FECHA"] + [f"{zona} ozono" for zona in zonas])
df_imeca
```

```python
df_imeca.info()
```

```python
df_imeca.Hora = pd.to_timedelta(df_imeca.Hora, unit="h")
df_imeca.Hora
```

```python
df_imeca["Fecha-Hora"] = df_imeca.Fecha + df_imeca.Hora
df_imeca["Fecha-Hora"]
```

```python
df_imeca = df_imeca.drop(columns=["Fecha", "Hora"])
df_imeca
```

```python
#Deteccion de valores faltantes
df_imeca = df_imeca.replace(-99, np.nan)
df_imeca.info()
```

```python
missing_values = df_imeca.isnull().sum()
missing_percentage = (missing_values / len(df_imeca)) * 100

# Filter to show only columns with missing values
missing_info = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
}).sort_values(by='Missing Percentage', ascending=False)

display(missing_info[missing_info['Missing Count'] > 0])
```

```python
#PONIENDO FECHA-HORA COMO LA PRIMERA COLUMNA
df_imeca = df_imeca[["Fecha-Hora"] + [col for col in df_imeca.columns if col != "Fecha-Hora"]]
df_imeca
```
